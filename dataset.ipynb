```python
import pandas as pd

file_path = r"C:\Users\a21ma\OneDrive\Desktop\Datahack\Datahack_Tensionflow\LawYantra Dataset - lawyer_summaries.csv"  
df = pd.read_csv(file_path)

information_list = []
for i in range(0, 1999):
    information_list.append(df.at[i,'Information'])

# print(information_list[1998])
```

```python
def separate_data(text):
    data_parts = text.split(". ")
    data_dict = {i: part.strip() for i, part in enumerate(data_parts)}
    return data_dict

feature_df = pd.DataFrame()

for i in range(len(information_list)):
    input_text = information_list[i]

    result = separate_data(input_text)

    # for key, value in result.items():
    #     print(f"Part {key}: {value}")
    
    feature_df_temp = (pd.DataFrame.from_dict(result, orient='index', columns=['Information']))
    feature_df_temp = feature_df_temp.T
    # feature_df_temp.index = [df.at[i, 'Lawyer Names']]
    # feature_df_temp = feature_df_temp.rename(index=df.at[i,'Lawyer Names'])
    feature_df = pd.concat([feature_df,feature_df_temp])

Lawyer_Names = []
for i in range(0, 1999):
    Lawyer_Names.append(df.at[i, 'Lawyer Names'])

feature_df.insert(0, "Lawyer Names", Lawyer_Names)
print(feature_df)
```

```python
feature_df.to_csv(r'C:\Users\a21ma\OneDrive\Desktop\Datahack\Datahack_Tensionflow\feature_df.csv', index = False)
```
