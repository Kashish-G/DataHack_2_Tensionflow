```python
import pandas as pd
```

```python
df1 = pd.read_csv('merged_df.csv')
df2 = pd.read_csv('merged_df2.csv')
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.sort_values(by='Lawyer Names', inplace=True, ignore_index=True)
df.drop(['First_Name'], axis=1, inplace=True)
df
```

```python
# df.to_csv('merged_df3.csv')
```

```python
df.drop_duplicates(subset=['Lawyer Names'], inplace=True, ignore_index=True)
df
```

```python
df['Gender'] = df['Gender'].replace({'female': 'Female', 'male': 'Male'})
df
```

```python
df.to_csv('final.csv')
```
