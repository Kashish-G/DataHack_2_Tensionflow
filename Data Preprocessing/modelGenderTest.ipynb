```python
import pandas as pd
import pickle
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
test = pd.read_csv(r"C:\Users\a21ma\OneDrive\Desktop\Datahack\DataHack_2_Tensionflow\no_duplicates_after_api.csv")
data = pd.read_csv(r'C:\Users\a21ma\OneDrive\Desktop\Datahack\DataHack_2_Tensionflow\Gender_Data.csv')
df = pd.DataFrame(data)
```

```python
names = test[pd.isna(test['Gender'])]
```

```python
names
```

```python
with open('model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
test_names = names['First_Name'].tolist()
test_names
```

```python
with open('tokenizer.pkl', 'rb') as file:
    tokenizer = pickle.load(file)
```

```python
# tokenizer = Tokenizer(char_level=True)  # Character-level tokenizer
X = tokenizer.texts_to_sequences(df['Name'])
X = pad_sequences(X)
```

```python
test_sequences = tokenizer.texts_to_sequences(test_names)
test_sequences = pad_sequences(test_sequences, maxlen=X.shape[1])

gender_mapping = {'male': 0, 'female': 1, 'unknown': 2}

predictions = loaded_model.predict(test_sequences)
print(predictions)
predicted_genders = [list(gender_mapping.keys())[np.argmax(prediction)] for prediction in predictions]
print(predicted_genders)
```

```python
names=names.drop("Gender",axis=1)
```

```python
names['Gender'] = predicted_genders
names
```

```python
test=test.dropna()
test
```

```python
test=pd.concat([test,names])
test
```
