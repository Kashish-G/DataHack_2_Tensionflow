```python
import pdfplumber
import openai
import time

response1 = []

def extract_text_page_by_page(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        for page_number in range(len(pdf.pages)):
            page = pdf.pages[page_number]
            page_text = page.extract_text()
            
            prompt1 = f'''
                Form a summary of the case mentioned in the text.
                {page_text}
                '''
            
            response = openai.Completion.create(
                model="text-davinci-003",
                prompt=prompt1,
                temperature=0.75,
                max_tokens=500,
                top_p=0.75,
                frequency_penalty=0.6,
                presence_penalty=0.8
            )
            
            summary = response.choices[0].text.strip()
            response1.append(summary)
            
            print(f"Page {page_number + 1} Summary:")
            print(summary)
            print("\n" + "="*50 + "\n")
            
            if (page_number + 1) % 3 == 0:
                print("Waiting for 1 minute...")
                time.sleep(60)  # Wait for 1 minute

pdf_path = "CASE OF ANAGNOSTAKIS v. GREECE.pdf"
extract_text_page_by_page(pdf_path)


```
